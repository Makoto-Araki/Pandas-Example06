```python
# ライブラリのインポート
import numpy as np

# 列(取引日)、行(商品名)のグループで取引量を集計(集計関数は合計を使用)
frame2.pivot_table(values='取引量', index='商品名', columns='取引日', aggfunc=np.sum)
```
